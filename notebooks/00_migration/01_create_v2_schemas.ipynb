```python
# ===========================================
# Notebook Name:
# 01_create_v2_schemas
#
# Purpose:
# Ensure pokemon.bronze / silver / gold / ops
# schemas exist. Idempotent.
# ===========================================
```

```python
CATALOG_NAME = "pokemon"

SCHEMAS = ["bronze", "silver", "gold", "ops"]
```

```python
for schema_name in SCHEMAS:
    spark.sql(
        f"CREATE SCHEMA IF NOT EXISTS "
        f"{CATALOG_NAME}.{schema_name}"
    )
    print(
        "Ensured schema:",
        f"{CATALOG_NAME}.{schema_name}",
    )
```

```python
display(
    spark.sql(
        f"SHOW SCHEMAS IN {CATALOG_NAME}"
    )
)
```
